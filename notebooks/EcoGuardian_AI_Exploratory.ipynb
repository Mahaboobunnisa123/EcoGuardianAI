# EcoGuardian AI - Exploratory Data Analysis

This notebook demonstrates initial analysis on environmental sensor data and visualizes key features related to microplastic contamination risk.


```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/epa_data_processed_CA.csv')

# Display basic info
print(df.info())
print(df.head())
```

```python
# Visualize distributions of key features
features = ['temp_c', 'ph', 'turbidity_ntu', 'dissolved_o2_mg_l', 'flow_rate_cfs']

for feature in features:
    plt.figure(figsize=(8,4))
    sns.histplot(df[feature].dropna(), bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```
